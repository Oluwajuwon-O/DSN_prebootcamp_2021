## how to create subplots

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(-5,5,20)
y = x**2 + 1
```

```python
# first define the size of the image/object
fig = plt.figure(figsize=(10,4))
'10 -> length'
'4 -> breadth'

# plt.subplots(nrows,ncols,plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(y,x,'g--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# seaborn comes with in-built dataset
sns.get_dataset_names()
```

```python
df = sns.load_dataset('mpg')
df.columns
```

```python
df.info()
```

```python
sns.distplot(df['mpg'])
plt.show()
```

```python
sns.distplot(df['mpg'], kde = False, bins = 20)
```

```python
sns.scatterplot(x='mpg', y='horsepower', data=df)
plt.show()
```

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', data = df, hue = 'cylinders')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x = 'mpg', y = 'horsepower', data = df, hue = 'cylinders', size = 'weight')
plt.show()
plt.savefig('scatter plot')
```

```python
sns.jointplot(x='weight', y='mpg',data=df)
```

```python
sns.jointplot(x='weight', y='mpg',data=df, kind = 'hex')
```

```python
sns.jointplot(x='weight', y='mpg',data=df, kind = 'reg')
```

# heatmap

```python
df_corr = df.corr()
df_corr
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(df_corr, cmap = 'coolwarm', annot = True)
# annot shows the number on the boxes
```

# Facet Grid

```python
df['origin'].value_counts()
```

```python
g = sns.FacetGrid(df, row = 'origin')
g = g.map(plt.scatter, 'mpg','weight') # apply a plotting function to each facet's subset of the data
```

```python
df['cylinders'].value_counts()
```

```python
g = sns.FacetGrid(df, col = 'cylinders')
g = g.map(plt.scatter, 'mpg','weight')
```

```python
g = sns.FacetGrid(df, row = 'origin', col = 'cylinders')
g = g.map(plt.scatter, 'mpg','weight')
```

```python
g = sns.FacetGrid(df, row = 'origin', col = 'cylinders', hue = 'origin')
g = g.map(plt.scatter, 'mpg','weight')
```
